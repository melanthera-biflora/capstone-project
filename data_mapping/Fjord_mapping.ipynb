```python
import numpy as np
import pandas as pd
```

#Concat the data

```python
band_value_germany = pd.read_csv('band_value_germany.csv')
band_value_germany_latest = pd.read_csv('band_value_germany_latest.csv')
salinity_df = pd.read_csv('salinity_german_data.csv')
```

```python
# concat band_value_germany and band_value_germany_latest

import pandas as pd
band_df = pd.concat([band_value_germany, band_value_germany_latest], ignore_index=True)
```

```python
band_df.info()
```

```python
salinity_df.info()
```

```python
# band_df and salinity_df drop Unnamed: 0

band_df = band_df.drop('Unnamed: 0', axis=1)
salinity_df = salinity_df.drop('Unnamed: 0', axis=1)
```

```python
band_df['Date'] = pd.to_datetime(band_df['Date'])
salinity_df['Date'] = pd.to_datetime(salinity_df['Date'])
```

```python
#  salinity_df merge with band_df on date, latitude, longitude, station

import pandas as pd
merged_df = pd.merge(salinity_df, band_df, on=['Date', 'Latitude', 'Longitude', 'Station'], how='left')
```
