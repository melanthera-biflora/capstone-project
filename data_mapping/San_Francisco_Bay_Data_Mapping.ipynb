#Concat and load the band data

```python
#  concat sfb_band_data_20020910, 20040518, 20061017, 20080916, 20090922, 20100330, 20011018

import numpy as np
import pandas as pd

# Assuming the files are in the same directory as your notebook
# and are named as sfb_band_data_YYYYMMDD.csv
file_names = ['sfb_band_data_20020910.csv', 'sfb_band_data_20040518.csv',
              'sfb_band_data_20061017.csv', 'sfb_band_data_20080916.csv',
              'sfb_band_data_20090922.csv', 'sfb_band_data_20100330.csv',
              'sfb_band_data_20111018.csv']

# Initialize an empty list to store dataframes
dfs = []

for file_name in file_names:
    try:
        # Read each CSV file into a pandas DataFrame
        df = pd.read_csv(file_name)
        dfs.append(df)
    except FileNotFoundError:
        print(f"Warning: File '{file_name}' not found. Skipping.")

# Concatenate all DataFrames in the list
if dfs:
  combined_df = pd.concat(dfs, ignore_index=True)
  print(combined_df)
else:
  print("No valid data files found.")
```
