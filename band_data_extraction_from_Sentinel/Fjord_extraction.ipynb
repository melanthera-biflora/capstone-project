```python
#Function to change lat long to bbox
def latlon_to_bbox(lat, lon, half_side_in_km):
    """
    This function returns the bounding box for a given latitude and longitude
    with the specified half-side distance in kilometers.
    """
    import math

    # Earth's radius in kilometers
    R = 6371

    # Coordinate offsets in radians
    half_side_in_rad = half_side_in_km / R

    lat_rad = math.radians(lat)
    lon_rad = math.radians(lon)

    # Latitude bounds
    min_lat = lat_rad - half_side_in_rad
    max_lat = lat_rad + half_side_in_rad

    # Constrain latitude bounds to be within -90 and 90 degrees
    min_lat = max(min_lat, math.radians(-90))
    max_lat = min(max_lat, math.radians(90))

    # Longitude bounds (longitude gets smaller when latitude increases)
    min_lon = lon_rad - half_side_in_rad / math.cos(lat_rad)
    max_lon = lon_rad + half_side_in_rad / math.cos(lat_rad)

    # Convert bounds back to degrees
    min_lat = math.degrees(min_lat)
    max_lat = math.degrees(max_lat)
    min_lon = math.degrees(min_lon)
    max_lon = math.degrees(max_lon)

    return min_lat, min_lon, max_lat, max_lon

```

```python
# function to return a 1-day time range from the given time

from datetime import datetime, timedelta

def get_one_day_time_range(given_time):
  """
  This function returns the start and end time of a 1-day time range from the given time.
  """
  start_time = given_time
  end_time = given_time + timedelta(days=1)
  return start_time, end_time
```

```python
import datetime
import os

import matplotlib.pyplot as plt
import numpy as np

from sentinelhub import (
    CRS,
    BBox,
    DataCollection,
    DownloadRequest,
    MimeType,
    MosaickingOrder,
    SentinelHubDownloadClient,
    SentinelHubRequest,
    bbox_to_dimensions,
)
```

```python
def band_extraction(min_lat, min_lon, max_lat, max_lon,start_time, end_time):

    evalscript_all_bands = """
      //VERSION=3
      function setup() {
          return {
              input: [
      { datasource: "md", bands: ["B01", "B02", "B03", "B04", "B05", "B06","B07"], unit: "REFLECTANCE" },
      { datasource: "l2a", bands: ["B01","B02","B03","B04","B05","B06","B07","B08","B8A","B09","B11","B12"], unit: "REFLECTANCE" }
    ],
              output: {
                  bands: 19,
                  sampleType: "INT16"
              }
          };
      }

      function evaluatePixel(sample) {
          var modis = sample.md[0];
          var sentinel = sample.l2a[0];
          return [modis.B01,
                  modis.B02,
                  modis.B03,
                  modis.B04,
                  modis.B05,
                  modis.B06,
                  modis.B07,
                  sentinel.B01,
                  sentinel.B02,
                  sentinel.B03,
                  sentinel.B04,
                  sentinel.B05,
                  sentinel.B06,
                  sentinel.B07,
                  sentinel.B08,
                  sentinel.B8A,
                  sentinel.B09,
                  sentinel.B11,
                  sentinel.B12,
                  ];
      }
  """
    resolution = 60
    betsiboka_bbox = BBox(bbox= (min_lat, min_lon, max_lat, max_lon), crs="EPSG:4326")
    betsiboka_size = bbox_to_dimensions(betsiboka_bbox, resolution=resolution)

    request_all_bands = SentinelHubRequest(
    evalscript=evalscript_all_bands,
    input_data=[
        SentinelHubRequest.input_data(
            data_collection=DataCollection.MODIS,
            identifier="md",
            time_interval=(str(start_time),str(end_time)),
            maxcc=0.2
        ),
        SentinelHubRequest.input_data(
            data_collection=DataCollection.SENTINEL2_L2A,
            identifier="l2a",
            time_interval=(str(start_time),str(end_time)),
            maxcc=0.2
        )
    ],
    responses=[SentinelHubRequest.output_response("default", MimeType.TIFF)],
    bbox=betsiboka_bbox,
    size=betsiboka_size,
    config=config,
)
    all_bands_data = request_all_bands.get_data()
    #plot_image(all_bands_data[0][:, :, 12], factor=3.5 / 1e4, vmax=1)
    #print(all_bands_data[0])
    average_values = np.mean(all_bands_data[0], axis=(0, 1))
    band_values = average_values.tolist()
    return band_values
```

```python
df = pd.DataFrame(columns=['Date','Station','Lat','Long','MODIS_B01','MODIS_B02','MODIS_B03', 'MODIS_B04','MODIS_B05','MODIS_B06', 'MODIS_B07',
                           'SENTINEL_B01','SENTINEL_B02','SENTINEL_B03', 'SENTINEL_B04','SENTINEL_B05','SENTINEL_B06', 'SENTINEL_B07', 'SENTINEL_B08', 'SENTINEL_B8A','SENTINEL_B09','SENTINEL_B11', 'SENTINEL_B12'])
```

```python
for date, station_number, lat, lon in lat_long:
  start_time, end_time = get_one_day_time_range(date)
  min_lon, min_lat, max_lon, max_lat = latlon_to_bbox(lat, lon, 5)
  band_values = band_extraction(min_lat, min_lon, max_lat, max_lon,start_time, end_time)
  print(band_values)
  if sum(band_values)!=0:
    value_list = [date, station_number, lat, lon] + band_values
    print(value_list)
    df.loc[len(df)] = value_list
```
